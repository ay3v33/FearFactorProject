```python
from transformers import pipeline

# Use a text2text-generation pipeline
chatbotFlan = pipeline("text2text-generation", model="google/flan-t5-large")
#chatbotFB = pipeline("text2text-generation", model="facebook/bart-large-cnn") summarizer
```

```python
import pandas as pd

df_short = pd.read_json("../data/MILDEC_short_out.jsonl", lines=True)
short_texts = df_short["text"].tolist()

df_long = pd.read_json("../data/MILDEC_long_out.jsonl", lines=True)
long_texts = df_long["text"].tolist()
```

```python
factors = ["fear", "stress", "morale", "trust_in_authorities"]

def fillPrompt(text_to_analyze, factor):
    return f"""
    You are trying to analyze how a person would feel given they read a certain text.
    Read the text and assign a value from 1 to 3 for how much {factor} this would induce.

    Text: "{text_to_analyze}"

    """

def getValues(text_to_analyze, factors):
    for factor in factors:
        prompt = fillPrompt(text_to_analyze, factor)
        response = chatbotFlan(prompt)[0]["generated_text"]
        print(response)
    
    
```

```python
getValues(long_texts[30], factors)
```

```python
getValues(short_texts[1000], factors)
```
